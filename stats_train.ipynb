`````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')

# Генерация синтетических данных временного ряда
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', end='2023-12-31', freq='D')
n = len(dates)

# Создаем временной ряд с трендом, сезонностью и шумом
trend = 0.05 * np.arange(n)  # линейный тренд
seasonal = 10 * np.sin(2 * np.pi * np.arange(n) / 365)  # годовая сезонность
noise = np.random.normal(0, 2, n)  # случайный шум

# Комбинируем компоненты
time_series = 100 + trend + seasonal + noise

# Создаем DataFrame
df_time = pd.DataFrame({
    'date': dates,
    'value': time_series,
    'day_of_week': dates.dayofweek,
    'month': dates.month,
    'quarter': dates.quarter
})

df_time.set_index('date', inplace=True)

print("📈 СГЕНЕРИРОВАННЫЕ ДАННЫЕ ВРЕМЕННОГО РЯДА:")

print(f"\nРазмер данных: {df_time.shape}")
df_time.head()
`````

`````python
# Завтра как сегоняд
test['naive_forecast'] = train['value'].iloc[-1]

# Среднее щна последние 30 д
last_30_avg = train['value'].tail(30).mean()
test['avg_forecast'] = last_30_avg

# С учетом тренда
last_value = train['value'].iloc[-1]
trend= (train['value'].iloc[-1] - train['value'].iloc[-30]) / 30
test['trand_forecast'] = last_value + trend * np.arange(1, len(test)+1)
`````
